#Get load and generation for any time from CAISO

```python
# Import main module
from __future__ import division
import getData
from pyiso import client_factory

# Imports useful for graphics
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style("whitegrid")
seaborn.despine()
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

saved = []
client = client_factory('CAISO')
```

```python
# Input data
year = 2015
month = 6
day = 21

result = getData.get_daily_data(year, month, day, client)
```

```python
# Plot data
plt.figure(figsize=(12, 14), dpi=80)
nbPlot = 410
for category in result:
    if not category == 'time':
        nbPlot += 1
        plt.subplot(nbPlot)
        plt.plot(result['time'], result[category], label=category)
        plt.xlabel('Time')
        plt.ylabel('Power (MW)')
        plt.legend()
plt.show()
```
